```python
import pandas as pd

# Load the LendingClub loan data (CSV file)
loan_df = pd.read_csv(r"C:\Users\rudhi\Downloads\IITM Hackathon\Lending Club\loan.csv")

# Checking for missing data
missing_data = loan_df.isnull().sum()

# For numerical features, fill missing values with the median
numerical_columns = loan_df.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_columns:
    loan_df[col].fillna(loan_df[col].median(), inplace=True)

# For categorical features, fill missing values with 'Unknown'
categorical_columns = loan_df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    loan_df[col].fillna('Unknown', inplace=True)

# Verify if there are any remaining missing values
missing_data_after = loan_df.isnull().sum()

# Display the first few rows of the cleaned data
loan_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Select only numeric columns
numeric_cols = loan_df.select_dtypes(include=['float64', 'int64']).columns

# Initialize the scaler
scaler = MinMaxScaler()

# Apply the scaler to numeric columns
loan_df[numeric_cols] = scaler.fit_transform(loan_df[numeric_cols])
```
